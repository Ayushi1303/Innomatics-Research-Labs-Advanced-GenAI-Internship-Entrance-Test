```python
##Smart Parking Lot Management System
def smart_parking_system(capacity, logs):
    current_parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1

        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

        # Alert if capacity exceeded
        if current_parked > capacity:
            print("ALERT: Parking capacity exceeded!")

    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)

    if current_parked < capacity:
        print("Parking Status: Available")
    else:
        print("Parking Status: Full")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

```python
##Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    # Base speed in km/h
    base_speed = 40

    # Calculate base time (in minutes)
    base_time = (distance / base_speed) * 60

    # Traffic delay
    if traffic == "High":
        base_time += 20
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "Low":
        base_time += 5

    # Weather delay
    if weather == "Rainy":
        base_time += 15
    elif weather == "Stormy":
        base_time += 25
    elif weather == "Clear":
        base_time += 0

    return round(base_time)


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

eta = estimate_delivery_time(distance, traffic, weather)
print("Estimated Delivery Time:", eta, "minutes")
```

```python
##Movie Theatre Seat Occupancy Analyzer
def analyze_seat_booking(total_seats, booked_seats):
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    elif occupancy >= 50:
        print("Show Status: Filling Fast")
    else:
        print("Show Status: Seats Available")

    # Suggest additional show
    if occupancy >= 90:
        print("Suggestion: Consider opening an additional show.")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

analyze_seat_booking(total_seats, booked_seats)
```

```python
##Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total = 0

    # Calculate total using loop
    for reading in cpu_readings:
        total += reading

    # Calculate average
    average_cpu = total / len(cpu_readings)

    print("Average CPU Load:", round(average_cpu), "%")

    # Classification
    if average_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= average_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
##Smart Classroom Resource Usage Monitor
def analyze_resource_usage(resource_usage):
    overused_resources = []
    threshold = 8  # hours limit

    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused_resources.append(resource)

    # Display results
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

analyze_resource_usage(resource_usage)
```

```python
##Online Event Registration Capacity Controller
def manage_event_registrations(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    # Loop through registrations
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
event_capacity = 100
registrations = 105

manage_event_registrations(event_capacity, registrations)
```
