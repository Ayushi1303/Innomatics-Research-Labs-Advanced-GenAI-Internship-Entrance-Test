```python
employees = { "Ravi": 75000, "Anita": 68000, "Kiran": 72000 }

top_employee=""
max_salary=0

for name in employees:
  max_salary=max(max_salary,employees[name])
  if max_salary==employees[name]:
    top_employee=name
print("Highest Salary:",top_employee,"-",max_salary)
```

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for status in attendance:
  if status == "P":
    present_count+=1


total_days = len(attendance)

percentage = (present_count / total_days)* 100

print("Attendance percentage:", percentage)
```

```python
text="pythonp"

char_count={}

for ch in text:
  if ch in char_count:
    char_count[ch]+=1
  else:
    char_count[ch]=1
print(char_count)
```
