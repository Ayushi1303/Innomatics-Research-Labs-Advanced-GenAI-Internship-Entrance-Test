```python
#Social Media – Post Engagement Analyzer
def analyze_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    print("Total Likes:", total_likes)

    # Check post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example usage
likes_data = [200, 300, 150, 400]   # Sample like counts
analyze_engagement(likes_data)
```

```python
## Healthcare – Medicine Stock Alert System
def monitor_stock(stock):
    print("Medicine Stock:", stock)

    # Conditional logic to check stock level
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


# Example usage
medicine_stock = 6
monitor_stock(medicine_stock)
```

```python
##Agriculture – Rainfall Adequacy Checker
def analyze_rainfall(rainfall_data, required_level):
    total = 0

    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total += rain

    # Calculate average rainfall
    average = total / len(rainfall_data)

    print("Average Rainfall:", int(average))

    # Check rainfall status
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example usage
seasonal_rainfall = [70, 75, 68, 80, 67]  # Sample data
required_rainfall_level = 70

analyze_rainfall(seasonal_rainfall, required_rainfall_level)
```

```python
##Social Media – Duplicate Account Detection
def detect_duplicate_usernames(usernames):
    unique_usernames = set(usernames)

    # If length differs, duplicates exist
    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
        print("Status: Flag accounts")
    else:
        print("Duplicate Accounts Found: No")
        print("Status: Accounts are unique")


# Example usage
username_list = ["john123", "alice22", "john123", "mark45"]

detect_duplicate_usernames(username_list)
```

```python
##Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    premium_crops = []

    # List filtering logic
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1500, 2500, 1800, 3200, 2000]
filter_premium_crops(crop_prices)
```

```python
##System Monitoring – Application Health Checker
def evaluate_application_health(error_count):
    print("Error Count:", error_count)

    # Conditional statements to evaluate system status
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example usage
errors = 7
evaluate_application_health(errors)
```

```python
##Banking – Daily Transaction Limit Checker
def validate_transaction(amount):
    daily_limit = 50000

    print("Transaction Amount:", amount)

    # Conditional logic to validate transaction
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example usage
transaction_amount = 60000
validate_transaction(transaction_amount)
```

```python
##E-Learning – Student Attendance Eligibility System
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Loop to count present days
    for status in attendance_list:
        if status == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / total_days) * 100

    print("Attendance Percentage:", percentage)

    # Check eligibility
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example usage
attendance_data = ["P", "P", "A", "P", "P"]  # 4 present out of 5 days
check_exam_eligibility(attendance_data)
```

```python
##Smart Electricity Bill Analyzer
def calculate_electricity_bill(units):
    bill = 0

    # Slab-wise billing calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status


# Example usage
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Classification:", status)
```
